#### 2. На основании условия (1) или «правила шести сигм» определить экстремальные наблюдения в выборке. Множитель в (1) подобрать таким образом, чтобы экстремальными признавалось не более 5-8% наблюдений

```python
for col in df.columns:
    x_med = df[col].median()
    x_mad = abs(df[col] - x_med).median()
    emissions = []
    coef = 1
    while True:
        for i in range(df[col].size):
            if df[col][i] < (x_med - coef*x_mad) or df[col][i] > (x_med + coef*x_mad):
                emissions.append(df[col][i])    
        pr = len(emissions) / df[col].size
        if 0.05 <= pr <= 0.08:
            print(f'{col}: percent={pr} coefficient={coef}')
            break
        coef += 0.1
        emissions = []
        
```

```python
k = []
for col in df.columns:
    emissions = []
    x_mean = df[col].mean()
    x_std = df[col].std()
    for i in range(df[col].size):
        if abs(df[col][i] - x_mean) > 6*x_std:
            emissions.append(df[col][i])
    k.append(len(emissions))
print(k)
```

### 5. КЛАСТЕРНЫЙ АНАЛИЗ ДАННЫХ

#### 1. Осуществить классификацию наблюдений с использованием алгоритма кластерного анализа («k-средних»). Количество классов задать равным четырём.

```python
n_clusters = 4
quantils = []
step = 1 / n_clusters
for i in range(1, n_clusters + 1):
    quantils.append(integral_indicator_of_creditworthiness.quantile(step * i))
quantils
```

```python
def get_num_cluster(x):
    for n, i in enumerate(quantils):
        if x <= i:
            return n
    return -1
```

```python
y = integral_indicator_of_creditworthiness.map(get_num_cluster)
y.unique()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4, random_state=1)

kmeans.fit(X)

# Visualizing using PCA
pca = PCA(n_components=2)
res = pca.fit_transform(X)
plt.figure(figsize=(10,6))
plt.scatter(res[:,0], res[:,1], c=y, s=50, cmap='jet')
plt.title('PCA')
```

```python
kmeans1 = KMeans(n_clusters=4, random_state=1)

kmeans1.fit(df_norm)

# Visualizing using PCA
pca = PCA(n_components=2)
res = pca.fit_transform(df_norm)
plt.figure(figsize=(10,6))
plt.scatter(res[:,0], res[:,1], c=kmeans1.labels_, s=50, cmap='jet')
plt.title('PCA')
```

### 6. ДИСКРИМИНАНТНЫЙ АНАЛИЗ ДАННЫХ

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
clf = LinearDiscriminantAnalysis()
clf.fit(X, y)
```

```python
auc_sum = {}
for i in y['label'].unique():
    print('Class', i)
    pred = clf.predict(X[y['label'] == i])
    auc_sum[str(i)] = accuracy_score(y[y['label'] == i], pred)
    print('Percentage of correct answers:',auc_sum[str(i)])
    unique, counts = np.unique(pred, return_counts=True)
    percent = [round(i / counts.sum(), 2) * 100 for i in counts]
    print('Class, predicted affiliation, Percentage of correct answers:')
    print(np.asarray((unique, counts, percent)).T)
    print('Number of all observations:', X[y['label'] == i].count()[0])
    print()
```

```python
np.mean([i[1] for i in auc_sum.items()])
```

Согласно этой информации 91% исходных наблюдений классифицированы верно. Наиболее высокая точность достигнута при классификации предприятий, относящихся к 0 классу платежеспособности (~97%), наименьшая – предприятий из 1 класса(~80).

Ниже приведены коэффициенты классифицирующих функций, использующихся для предсказания класса, которому принадлежит новое наблюдения. Правило классификации следующее: наблюдение принадлежит классу кредитоспособности i , для классифицирующей функции которого d i получено наибольшее значение.

```python
pd.DataFrame(clf.coef_.T, columns=['d1', 'd2', 'd3', 'd4'], index=X.columns)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
clf = LinearDiscriminantAnalysis()
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
for i in range(4):
    correct = 0
    incorrect = 0
    print('Class', i)
    for j, value in enumerate(y_test.values):
        if value[0] == i and y_pred[j] == i:
            correct += 1
        elif value[0] == i:
            incorrect += 1
    print('Number of correct answers:', correct, '\nNumber of correct answers:', incorrect,\
          '\nPercentage of incorrect answers:', round(incorrect / (incorrect + correct), 2) * 100)
```

```python
print('Accuracy:', accuracy_score(y_test, y_pred))
```

Приведены коэффициенты классифицирующих функций, полученные при реализации дискриминантного анализа к сокращенной обучающей выборке за счет выделения экзаменационной выборки.

```python
pd.DataFrame(clf.coef_.T, columns=['d1', 'd2', 'd3', 'd4'], index=X.columns)
```
